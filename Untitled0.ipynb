```python
import numpy as np
a1=30.0
a2=100.0
b1=0.01
b2=0.04
N1=50 #number of abonents on stage 1
N2=230 #number of abonents on stage 2
C1=0.5 #variable cost on stage 1
C2=0.2 #variable cost on stage 2
max_M1=3 #maximum number of companies on 1 stage +1
max_M2=1001 #maximum number of companies on 2 stage +1
q1=np.zeros((max_M1,max_M2))
q2=np.zeros((max_M1,max_M2))
p1=np.zeros((max_M1,max_M2))
p2=np.zeros((max_M1,max_M2))
Q1=np.zeros((max_M1,max_M2))
Q2=np.zeros((max_M1,max_M2))
eta1=np.zeros((max_M1,max_M2))
eta2=np.zeros((max_M1,max_M2))
for M1 in range(1,3):
  for M2 in range(M1+1,max_M2):
    k=1-M1/M2
    k_sq=k*k
    div=4*b1*b2*M1*M2+(2*b2*M2*M2)*k_sq
    q1[M1,M2]=(2*b2*M2*k*a2-2*b2*M2*C1-((2*b2*b2*M2*M2+1)*k-2*b2*M2)*C2)/div
    q2[M1,M2]=((M2*k_sq+k+2*M1*b1)*a2-k*C1-(M2*(1+b2)*k_sq-k+2*b1*M1)*C2)/div
    Q1[M1,M2]=M1*q1[M1,M2]
    Q2[M1,M2]=M2*q2[M1,M2]
p1=a1-b1*M1*q1
p2=a2-b2*M2*q2
eta1=-b1*p1/(a1-b1*p1)
eta2=-b2*p2/(a2-b2*p2)



```

```python
M1=1
import matplotlib.pyplot as plt
plt.plot(q1[M1,M1:max_M2],q2[M1,M1:max_M2],'r-')
plt.plot(q1[M1+1,M1+1:max_M2],q2[M1+1,M1+1:max_M2],'b--')
plt.plot(q1[M1,M1+1],q2[M1,M1+1],'g^')
plt.plot(q1[M1+1,M1+2],q2[M1+1,M1+2],'g^')
plt.xlabel('q*_1')
plt.ylabel('q*_2')
plt.show()
```

```python
M1=1
import matplotlib.pyplot as plt
plt.plot(Q1[M1,M1:max_M2],Q2[M1,M1:max_M2],'r-')
plt.plot(Q1[M1+1,M1+1:max_M2],Q2[M1+1,M1+1:max_M2],'b--')
plt.plot(Q1[M1,M1+1],Q2[M1,M1+1],'g^')
plt.plot(Q1[M1+1,M1+2],Q2[M1+1,M1+2],'g^')
plt.xlabel('Q*_1')
plt.ylabel('Q*_2')
plt.show()


```
