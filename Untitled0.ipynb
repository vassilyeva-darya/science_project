```python
import numpy as np
a1=30.0
a2=100.0
b1=0.01
b2=0.5
N1=50 #number of abonents on stage 1
N2=230 #number of abonents on stage 2
c1=0.5 #variable cost on stage 1
c2=0.2 #variable cost on stage 2
c=1000000 #constant cost
max_M1=3 #maximum number of companies on 1 stage +1
max_M2=101 #maximum number of companies on 2 stage +1
k=np.zeros((max_M1,max_M2))
q1=np.zeros((max_M1,max_M2))
q2=np.zeros((max_M1,max_M2))
p1=np.zeros((max_M1,max_M2))
p2=np.zeros((max_M1,max_M2))
Q1=np.zeros((max_M1,max_M2))
Q2=np.zeros((max_M1,max_M2))
eta1=np.zeros((max_M1,max_M2))
eta2=np.zeros((max_M1,max_M2))
TPl=np.zeros((max_M1,max_M2))
TPf=np.zeros((max_M1,max_M2))
TP=np.zeros((max_M1,max_M2))
CS1=np.zeros((max_M1,max_M2))
CS2=np.zeros((max_M1,max_M2))
CS=np.zeros((max_M1,max_M2))
SS=np.zeros((max_M1,max_M2))
for M1 in range(1,3):
  for M2 in range(M1,max_M2):
    k[M1,M2]=1-M1/M2
    k_sq=k[M1,M2]*k[M1,M2]
    r=4*b1*b2*M1*M2+(2*b2*M2*M2-1)*k_sq
    q1[M1,M2]=((2*b2*M2+k[M1,M2])*a2-2*b2*M2*c1-((2*b2*b2*M2*M2+1)*k[M1,M2]-2*b2*M2)*c2)/r
    q2[M1,M2]=((M2*k_sq+k[M1,M2]+2*M1*b1)*a2-k[M1,M2]*c1-(M2*(1+b2)*k_sq-k[M1,M2]+2*b1*M1)*c2)/r
    Q1[M1,M2]=M1*q1[M1,M2]
    Q2[M1,M2]=M2*q2[M1,M2]
    p1[M1,M2]=a1-b1*M1*q1[M1,M2]
    p2[M1,M2]=a2-b2*M2*q2[M1,M2]
    eta1[M1,M2]=-b1*p1[M1,M2]/(a1-b1*p1[M1,M2])
    eta2[M1,M2]=-b2*p2[M1,M2]/(a2-b2*p2[M1,M2])
    TPl[M1,M2]=(p1[M1,M2]-c1)*q1[M1,M2]+(p2[M1,M2]-c2)*(q2[M1,M2]-q1[M1,M2])-c
    TPf[M1,M2]=(p2[M1,M2]-c2)*q2[M1,M2]-0.5*c
    TP[M1,M2]=M1*TPl[M1,M2]+(M2-M1)*TPf[M1,M2]
    CS1[M1,M2]=(a1-p1[M1,M2])*Q1[M1,M2]/2
    CS2[M1,M2]=(a2-p2[M1,M2])*Q2[M1,M2]/2
    CS[M1,M2]=CS1[M1,M2]+CS2[M1,M2]
    SS=TP[M1,M2]+CS[M1,M2]
    


```
